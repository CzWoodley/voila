# Data description and visualization

```python
# descriptive statistics
health_factors_data_brazil.loc[:, ['under-five deaths ', 'Hepatitis B', 'Polio']].mean()
```

The mean under-five deaths in Brazil, between 2001 and 2015, is 77.25 deaths per 1000 population. 
Additionally, the mean percentage  immunization coverage among 1-year-olds of Hepatitis B is 96.2%, while that of Polio is 98.3%. 

```python
health_factors_data_brazil.loc[:, ['Hepatitis B', 'Polio']].mode()
```

Between 2001 and 2015, majority of the calendar years in Brazil had a 96% Hepatitis B immunization coverage among 1-year-olds, and a 99% Polio immunization coverage among 1-year-olds. 

```python
hepB_lineplot = alt.Chart(health_factors_data_brazil).mark_line().encode(
    y=alt.Y('Hepatitis B', scale=alt.Scale(domain=[90, 100])),
    x=alt.X('Year', scale=alt.Scale(domain=[2000, 2015])), 
    tooltip = ['Year', 'Hepatitis B'],
    ).properties(
    title="Percentage Hepatitis B immunization coverage among Brazilian 1-year-olds"
    )

polio_lineplot = alt.Chart(health_factors_data_brazil).mark_line().encode(
    y=alt.Y('Polio', scale=alt.Scale(domain=[95, 100])),
    x=alt.X('Year', scale=alt.Scale(domain=[2000, 2015])), 
    tooltip = ['Year', 'Polio'], 
    ).properties(
    title="Percentage Polio immunization coverage among Brazilian 1-year-olds"
    )

under_five_death_lineplot = alt.Chart(health_factors_data_brazil).mark_line().encode(
    y=alt.Y('under-five deaths '),
    x=alt.X('Year', scale=alt.Scale(domain=[2000, 2015])), 
    tooltip = ['Year', 'under-five deaths '],
    ).properties(
    title="Under-five deaths per 1000 population in Brazil"
    )

hepB_lineplot | polio_lineplot | under_five_death_lineplot
```

The graphs shows that there is a difference in trend between the hepatitis B - and the polio immunization coverage. 
The largest decrease in Hepatitis B immunization coverage is observed between 2000 and 2001, and the largest increase between 2002 and 2003. The Hepatitis B immunization coverage remains stagnant after 2012. 

<br>

On the other hand, Polio immunization coverage is stagnant between 2000 and 2010, followed by a large decrease between 2010 and 2012, and it slightly increases again in 2014.
